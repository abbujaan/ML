```python
#2
import pandas as pd
```

```python
data = pd.read_csv("/kaggle/input/temperatures-of-india/temperatures.csv")
```

```python
data
```

```python
x_train=data[['YEAR']]
y_train=data[['ANNUAL']]
```

```python
from sklearn.linear_model import LinearRegression 
model=LinearRegression()
model.fit(x_train,y_train)
y_predict=model.predict(x_train)
```

```python

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mean_squared_error(y_train,y_predict)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train, label ="actual", color="red")
plt.plot(x_train,y_predict,label ="predicted", color="blue")
plt.xlabel('year')
plt.ylabel('temp')
plt.title('annual prediction')
plt.legend()
```
